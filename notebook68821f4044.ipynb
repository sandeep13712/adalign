```python
def retrieve(query, k=5):
    """Retrieve top-k relevant ads for a query"""
    q_emb = embedder.encode([query], convert_to_numpy=True)
    distances, indices = index.search(q_emb, k)
    return [ads[i] for i in indices[0]]

# 4. Generate response with LLM
def generate_response(query, use_openai=False):
    retrieved_ads = retrieve(query)
    context = "\n".join(retrieved_ads)
    
    if use_openai:
        prompt = f"""
                    You are an ad recommender.
                    User query: {query}
                    Relevant ads:
                    {context}
                    
                    Provide a helpful recommendation to the user using the ads above.
                    """
        response = openai.ChatCompletion.create(
            model="gpt-4o-mini",
            messages=[{"role": "user", "content": prompt}]
        )
        return response["choices"][0]["message"]["content"]
    else:
        # fallback mode: just print retrieved ads
        return f"Based on your query '{query}', we recommend:\n- " + "\n- ".join(retrieved_ads)
```

```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
import openai

# 1. Load ads file
with open("/kaggle/working/ads.txt", "r", encoding="utf-8") as f:
    ads = [line.strip() for line in f if line.strip()]

# 2. Create embeddings
embedder = SentenceTransformer("all-MiniLM-L6-v2")
ad_embeddings = embedder.encode(ads, convert_to_numpy=True)

# 3. Store in FAISS index
d = ad_embeddings.shape[1]
index = faiss.IndexFlatL2(d)
index.add(ad_embeddings)

# ---- Try it ----
user_query = "I want new shoes for jogging"
print(generate_response(user_query, use_openai=False))

```

```python
# Install required libraries
!pip install torch torchvision pillow easyocr ultralytics

from ultralytics import YOLO
from PIL import Image
import easyocr

# ========== Load YOLOv8 model for object detection ==========
model = YOLO("yolov8n.pt")  # small version suitable for Kaggle

# ========== Initialize OCR reader ==========
reader = easyocr.Reader(['en'])

# ========== Function to extract structured content ==========
def extract_image_content(image_path):
    # Load image
    img = Image.open(image_path).convert("RGB")
    
    # 1. Object detection
    results = model.predict(source=image_path)
    objects = []
    for result in results:
        for box in result.boxes:
            class_id = int(box.cls[0])
            class_name = model.names[class_id]
            objects.append(class_name)
    
    # 2. OCR / text extraction
    ocr_results = reader.readtext(image_path)
    text_in_image = [text[1] for text in ocr_results] if ocr_results else []
    
    # 3. Simple color extraction (dominant color approximation)
    colors = img.convert('RGB').getcolors(maxcolors=1000000)
    colors_sorted = sorted(colors, key=lambda x: x[0], reverse=True)
    dominant_color = colors_sorted[0][1] if colors_sorted else None
    
    # Return structured content
    return {
        "objects": list(set(objects)),
        "dominant_color": dominant_color,
        "text_in_image": text_in_image
    }

# ========== Example usage ==========
image_path = "example_product.jpg"  # Replace with your image
content = extract_image_content(image_path)
print("Structured Image Content:", content)
```
